```python
SMA1 = 42
SMA2 = 252
```

```python
data['SMA1'] = data[symbol].rolling(SMA1).mean()
data['SMA2'] = data[symbol].rolling(SMA2).mean()
```

```python
data.plot(figsize=(10, 6))
```

```python
data.dropna(inplace=True)
```

```python
# Condition, True representation, False Representation
data['Position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
ax = data.plot(secondary_y='Position' ,figsize=(10,6))
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```

```python
# Shift goes back one t
data['Returns'] = np.log(data[symbol] / data[symbol].shift(1))
```

```python
# If you represent the conditional return as 1 and -1 you get the appropriate log return for
# Long/Short this way. The shift is to avoid foresight bias caused by 
# flipping the strategy intraday when the averages cross
data['Strategy'] = data['Position'].shift(1) * data['Returns']
```

```python
data.round(4).head()
```

```python
data.dropna(inplace=True)
```

```python
# Exponential of the sum of the log returns - total performance
np.exp(data[['Returns', 'Strategy']].sum())
```

```python
# vol target
data[['Returns' ,'Strategy']].std() * 252 ** 0.5
```

```python
# Graph benchmark vs. strat
ax = data[['Returns', 'Strategy']].cumsum(
                 ).apply(np.exp).plot(figsize=(10,6))
data['Position'].plot(ax=ax, secondary_y ='Position', style='--')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```

```python
from itertools import product
sma1 = range(20, 61, 4)
sma2 = range(180, 281, 10)
results = pd.DataFrame()
for SMA1, SMA2 in product(sma1, sma2):
    data = pd.DataFrame(raw[symbol])
    data.dropna(inplace=True)
    data['Returns'] = np.log(data[symbol] / data[symbol].shift(1))
    data['SMA1'] = data[symbol].rolling(SMA1).mean()
    data['SMA2'] = data[symbol].rolling(SMA2).mean()
    data.dropna(inplace=True)
    
    data['Position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
    data['Strategy'] = data['Position'].shift(1) * data['Returns']
    data.dropna(inplace=True)
    perf = np.exp(data[['Returns', 'Strategy']].sum())
    results = results.append(pd.DataFrame(
    {
        'SMA1' : SMA1, 'SMA2': SMA2,
        'MARKET' : perf['Returns'],
        'STRATEGY': perf['Strategy'],
        'OUT': perf['Strategy'] - perf['Returns']
    }, index=[0]), ignore_index=True)
    
```

```python
# Additionally, you can transform a Long/Short position into 1 or -1 and then multiple that vector by the
# returns vector to get the returns of a strat
```

```python
from sklearn import linear_model
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
C = 1
```

```python
models = {
    'log_reg': linear_model.LogisticRegression(C=C),
    'gauss_nb': GaussianNB(),
    'svm': SVC(C=C)
}
```

```python
def fit_models(data):
    mfit = {model: models[model].fit(data[cols_bin],
                                    data['direction'])
           for model in models.keys()}
    
```

```python
fit_models(data)
```

```python
def derive_positions(data):
    for model in models.keys():
        data['pos_' + model] = models[model].predict(data[cols_bin])
```

```python
derive_positions(data)
```

```python
# run every strat possible per security
# let rip whichever one backtests best
def evaluate_strats(data):
    global sel
    sel = []
    for model in models.keys():
                 col = 'strat_' + model
                 data[col] = data['pos_' + model] * data['returns']
                 sel.append(col)
    sel.insert(0, 'returns')
```

```python
evaluate_strats(data)
```

```python
sel.insert(1, 'strat_freq')
```

```python
data[sel].sum().apply(np.exp)
```

```python
# cumsum() is the sum of each cell and all cells before it along the axis specified
# In this case, we're adding up the returns of each strat, then apply()-ing np.exp to
# take the log-returns, which are wide sense stationary
data[sel].cumsum().apply(np.exp).plot(figsize=(10,6))
```
