```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
import numpy as np
np.random.seed(999)
y = np.random.standard_normal(20)
x = np.linspace(0, 12, 20) # last arg has to be same as number of y values
plt.plot(x, y)
```

```python
# lol just for fun plot a function
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
plt.xlim(-2,5)
plt.ylim(np.min(y.cumsum())-1, np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(20,12))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'gX')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Simple Plot')
```

```python
'''
cumsum is the performant way of normal samples for a Wiener process, because of the identity that a stochastic process
S(t) === SUM(0:t-1) + t and Python generates normally distributed sample draws efficiently
''' 
y = np.random.standard_normal((20, 2)).cumsum(axis=0) 
y[0:5]
```

```python
plt.figure(figsize=(20,12))
plt.plot(y, lw=1.5)
plt.plot(y, 'rX')
plt.xlabel('index')
plt.ylabel('value')
plt.title('2 Series plot')
```

```python
# Accessing individual columns of the series
plt.figure(figsize=(20,12))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0) # loc 0 is 'best' location: let Python decide
plt.xlabel('index')
plt.ylabel('value')
plt.title('Individual column label plot')
```

```python
# Two different y-axis

fig, ax1 = plt.subplots(figsize=(20,12))
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=2)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('Dual y-axis plot')
ax2 = ax1.twinx() # this actually generates a second subplot sharing the x axis of first one
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# Two separate plots
plt.figure(figsize=(20, 12))
plt.subplot(211) # This is a coordinate system (num rows, num columns, which plots)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Two Separate Plots')
plt.subplot(212) # So, this is (2 rows, 1 column, 2nd plot)
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
```

```python
# 2 different types of plots in a figure
plt.figure(figsize=(20,12))
plt.subplot(121) # 1 row, 2 columns, plot 1
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('1st Dataset') # 1 plt.title per subplot in this style
plt.subplot(122) # 1 row, 2 columns, plot 2
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd Data Set') # TODO refactor all of my long/short strat graphs with signals to a subplot(21*) style
```

```python
# Scatter plot
y = np.random.standard_normal((1000, 2)) # 1000 points in 2D space
plt.figure(figsize=(20,12))
# plt.scatter(y[:, 0], y[:, 1]) # Same thing as next line
plt.plot(y[:, 0], y[:, 1], 'r^') 
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# Scatter plotting a 3-dimension vector
c = np.random.randint(0, 10, len(y)) # vector of random integer between 0 and 9 of length y
plt.figure(figsize=(20,12))
plt.scatter(y[:, 0], y[:, 1], c=c, cmap='coolwarm', marker='o') # c kwarg is values for colormap
plt.colorbar()
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# Histogram
plt.figure(figsize=(20,12))
plt.hist(y, label=['1st', '2nd'], bins=25) # number of bucket to bin values into
plt.legend(loc=0) # best choice
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
'''
Full Params for Histogram:
plt.hist(x, bins=10, range=None, normed=False, weights=None, cumulative=False,
bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None,
log=False, color=None, label=None, stacked=False, hold=None, **kwargs)
'''
```

```python
# Histogram two draw series from a Gaussian dataset
plt.figure(figsize=(20,12))
plt.hist(y, label=['1st', '2nd'], color=['b', 'g'], stacked=True, bins=20, alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequecy')
plt.legend(loc=0)
plt.title('Histogram')
```

```python
# Boxplot
fig, ax = plt.subplots(figsize=(20,12))
plt.boxplot(y)
# set property for a a subplot on a plot - first arg which plot, 2nd arg a kwarg
plt.setp(ax, xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
'''
Some Boxplot Reminders:
Whiskers are 1.5 times 75th percentile val - 25th percentile val
Bottom of box is 25th
Top of box is 75th
Line on box is median
Circle are outliers
'''
```

```python
''' 
Now! 3d Plotting for vol surfaces
A Vol surface is generated with a matrix of strike prices and time-to-maturity values, whose implied volatility is the
3rd dimension
'''
strike = np.linspace(50, 150, 24)
ttm = np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike, ttm) # coordinate grid for the vol surface
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
'''
So this is a dummy function for the implied volatility of each option.
In real life, IV is computed iteratively with a bisection approach until sufficient accuracy is reached
'''
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,12))
ax = plt.gca(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True) 
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
```
