```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
import numpy as np
np.random.seed(999)
y = np.random.standard_normal(20)
x = np.linspace(0, 12, 20) # last arg has to be same as number of y values
plt.plot(x, y)
```

```python
# lol just for fun plot a function
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
plt.xlim(-2,5)
plt.ylim(np.min(y.cumsum())-1, np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(20,12))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'gX')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Simple Plot')
```

```python
'''
cumsum is the performant way of normal samples for a Wiener process, because of the identity that a stochastic process
S(t) === SUM(0:t-1) + t and Python generates normally distributed sample draws efficiently
''' 
y = np.random.standard_normal((20, 2)).cumsum(axis=0) 
y[0:5]
```

```python
plt.figure(figsize=(20,12))
plt.plot(y, lw=1.5)
plt.plot(y, 'rX')
plt.xlabel('index')
plt.ylabel('value')
plt.title('2 Series plot')
```

```python
# Accessing individual columns of the series
plt.figure(figsize=(20,12))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0) # loc 0 is 'best' location: let Python decide
plt.xlabel('index')
plt.ylabel('value')
plt.title('Individual column label plot')
```

```python
# Two different y-axis

fig, ax1 = plt.subplots(figsize=(20,12))
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=2)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('Dual y-axis plot')
ax2 = ax1.twinx() # this actually generates a second subplot sharing the x axis of first one
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# Two separate plots
plt.figure(figsize=(20, 12))
plt.subplot(211) # This is a coordinate system (num rows, num columns, which plots)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Two Separate Plots')
plt.subplot(212) # So, this is (2 rows, 1 column, 2nd plot)
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
```

```python
# 2 different types of plots in a figure
plt.figure(figsize=(20,12))
plt.subplot(121) # 1 row, 2 columns, plot 1
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('1st Dataset') # 1 plt.title per subplot in this style
```
