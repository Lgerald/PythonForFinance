```python
import pandas as pd
import numpy as np
import os
print(os.environ['IEX_TOKEN'])
```

```python
# Finance data time
from datetime import datetime
from iexfinance.stocks import get_historical_data

start = datetime(2019, 1, 1)
end = datetime(2019, 6, 15)
pd.options.display.float_format = '{:,.2f}'.format
df = pd.DataFrame(get_historical_data("AAPL", start, end, close_only=True)).transpose()
```

```python
def streaker(x, count):
    if(x > 1):
        count += 1
    else:
        count = 0
    return count
```

```python
df['log_returns'] = np.log(df['close']/df['close'].shift(1))
```

```python
df['ups'] = np.where(df['log_returns'] > 0, 1, 0)
df['downs'] = np.where(df['log_returns'] <= 0, 1, 0)
df['trend_up'] = np.cumsum(df['ups'])
df['trend_down'] = np.cumsum(df['downs'])
```

```python
df['streaks'] = df['log_returns'].apply(streaker(df['trend_count'][0]))
```
